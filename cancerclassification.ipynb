```python
df.drop(columns='diagnosis',inplace=True)
```

```python
X = df.drop(columns='benign_0_mal_1').values
y = df['benign_0_mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=100,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
model.fit(x=X_train,y=y_train,epochs=700,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
losses_early = pd.DataFrame(model.history.history)
```

```python
losses_early
```

```python
losses_early.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
model.fit(x=X_train,y=y_train,epochs=700,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
losses_dropout_early = pd.DataFrame(model.history.history)
```

```python
losses_dropout_early.plot()
```

```python
preds = model.predict(X_test)
```

```python
predictions = (preds > 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
